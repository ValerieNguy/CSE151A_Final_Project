#Scatter Plots

```python
print("Max Trip Distance:", df['trip_distance'].max())
print("Max Fare Amount:", df['fare_amount'].max())
```

```python
import matplotlib.pyplot as plt

# Set thresholds for filtering outliers
max_trip_distance = 50  # miles
max_fare_amount = 500  # dollars

# Filter data to remove negative values and outliers based on defined thresholds
df_filtered = df[(df['trip_distance'] >= 0) & (df['fare_amount'] >= 0) &
                 (df['trip_distance'] <= max_trip_distance) & (df['fare_amount'] <= max_fare_amount)]

# Plot the filtered data
plt.figure(figsize=(8, 6))
plt.scatter(df_filtered['trip_distance'], df_filtered['fare_amount'], alpha=0.5)
plt.xlabel("Trip Distance (miles)")
plt.ylabel("Fare Amount ($)")
plt.title("Fare Amount vs Trip Distance (Filtered)")
plt.show()
```

```python
# Set threshold for filtering outliers in fare_amount
max_fare_amount = 500  # dollars

# Filter data to remove negative values and cap fare_amount at $500
df_filtered = df[(df['fare_amount'] >= 0) & (df['fare_amount'] <= max_fare_amount)]

# Plot the filtered data
plt.figure(figsize=(10, 6))
plt.scatter(df_filtered['PULocationID'], df_filtered['fare_amount'], alpha=0.5, color='blue')
plt.xlabel("Pickup Location ID")
plt.ylabel("Fare Amount ($)")
plt.title("Fare Amount vs Pickup Location (Filtered)")
plt.show()
```

```python
# Display unique values of PULocationID
unique_pickup_ids = df_filtered['PULocationID'].unique()
print("Unique Pickup Location IDs:")
print(unique_pickup_ids)

# Display the count of each unique PULocationID
pickup_location_counts = df_filtered['PULocationID'].value_counts()
print("Pickup Location ID Counts:")
print(pickup_location_counts.head(10))  # Display the top 10 most frequent locations
```

```python
# Convert pickup and dropoff datetime columns to datetime format
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])

# Extract the hour of pickup
df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour

# Set threshold for filtering outliers in fare_amount
max_fare_amount = 500  # dollars

# Filter data to remove negative fare amounts and cap fare_amount at $500
df_filtered = df[(df['fare_amount'] >= 0) & (df['fare_amount'] <= max_fare_amount)]

# Plot the scatter plot of Pickup Hour vs Fare Amount
plt.figure(figsize=(10, 6))
plt.scatter(df_filtered['pickup_hour'], df_filtered['fare_amount'], alpha=0.5, color='green')
plt.xlabel("Pickup Hour")
plt.ylabel("Fare Amount ($)")
plt.title("Fare Amount vs Pickup Hour (Filtered)")
plt.show()
```

```python
# Filter data to remove negative congestion surcharges
df_filtered = df_filtered[df_filtered['congestion_surcharge'] >= 0]

# Plot the scatter plot with flipped axes
plt.figure(figsize=(10, 6))
plt.scatter(df_filtered['congestion_surcharge'], df_filtered['pickup_hour'], alpha=0.5, color='purple')
plt.xlabel("Congestion Surcharge ($)")
plt.ylabel("Pickup Hour")
plt.title("Pickup Hour vs Congestion Surcharge (Filtered)")
plt.show()
```

```python
# Plot the scatter plot of Pickup Hour vs Improvement Surcharge
df_filtered = df_filtered[df_filtered['improvement_surcharge'] >= 0]

# Plot the scatter plot with flipped axes
plt.figure(figsize=(10, 6))
plt.scatter(df_filtered['improvement_surcharge'], df_filtered['pickup_hour'], alpha=0.5, color='blue')
plt.xlabel("Improvement Surcharge ($)")
plt.ylabel("Pickup Hour")
plt.title("Pickup Hour vs Improvement Surcharge (Filtered)")
plt.show()
```
